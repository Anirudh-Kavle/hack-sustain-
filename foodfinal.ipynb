```python
# Define your categorization
def categorize_diet(entity):
    vegan_items = ['Apples', 'Bananas', 'Barley', 'Beet Sugar', 'Berries & Grapes', 'Brassicas', 'Cane Sugar',
                   'Cassava', 'Citrus Fruit', 'Coffee', 'Dark Chocolate', 'Groundnuts', 'Maize', 'Nuts', 'Oatmeal',
                   'Onions & Leeks', 'Other Fruit', 'Other Pulses', 'Other Vegetables', 'Peas', 'Potatoes',
                   'Root Vegetables', 'Soy milk', 'Tofu', 'Tomatoes', 'Wheat & Rye', 'Wine']
    vegetarian_items = ['Cheese', 'Eggs', 'Milk']
    meat_eater_items = ['Beef (beef herd)', 'Beef (dairy herd)', 'Fish (farmed)', 'Lamb & Mutton', 'Pig Meat',
                        'Poultry Meat', 'Prawns (farmed)', 'Rice']

    if entity in vegan_items:
        #return 'Vegan'
        return 1
    elif entity in vegetarian_items:
        #return 'Vegetarian'
        return 2
    elif entity in meat_eater_items:
        #return 'Meat Eater'
        return 3
    else:
        #return 'Uncategorized'
        return 4

# Apply the categorization function and create a new column
df['Diet'] = df['Entity'].apply(categorize_diet)
```

```python
df.rename(columns={'GHG emissions per kilogram (Poore & Nemecek, 2018)':'GHG'},inplace=True)
```

```python
def diet(x):
    if x=='Meat Eater':
        print("Mean of Non Veg GHG: ",(df.query("Diet<=3")["GHG"]).mean())
    elif x=='Vegetarian':
        print("Mean of Veg GHG: ",(df.query("Diet<=2")["GHG"]).mean())
    else:
        print("Mean of Vegan GHG: ",(df.query("Diet==1")["GHG"]).mean())
```

```python
diet("Meat Eater")
diet("Vegetarian")
diet("vegan")
```
